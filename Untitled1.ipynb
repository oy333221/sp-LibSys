```python
def crawl_book_info(isbn):
    url = f"https://www.googleapis.com/books/v1/volumes?q=isbn:{isbn}"
    try:
        response = requests.get(url, timeout=5)  # 設定 5 秒超時
        if response.status_code == 200 and response.json()['totalItems'] > 0:
            book = response.json()['items'][0]['volumeInfo']
            return {
                'title': book.get('title', '未知書名'),
                'author': ', '.join(book.get('authors', ['未知作者'])),
                'description': book.get('description', '無描述'),
                'product_link': f"https://books.google.com/books?isbn={isbn}"
            }
        else:
            print(f"ISBN {isbn} 未找到書籍資訊")
            return None
    except requests.exceptions.RequestException as e:
        print(f"網路錯誤爬取 ISBN {isbn}: {str(e)}")
        return None
    except Exception as e:
        print(f"其他錯誤爬取 ISBN {isbn}: {str(e)}")
        return None
```

```python
def update_pending_book(isbn):
    info = crawl_book_info(isbn)
    if info:
        try:
            supabase.table('pending_books').update({
                'title': info['title'],
                'author': info['author'],
                'description': info.get('description', '無描述'),
                'product_link': info['product_link']
            }).eq('isbn', isbn).execute()
            print(f"成功更新 ISBN {isbn} 的資訊")
            return True
        except Exception as e:
            print(f"更新 ISBN {isbn} 失敗: {str(e)}")
            return False
    return False
```

```python
def monitor_pending_books(max_iterations=5):
    """在 Jupyter 中監控 pending_books 表格"""
    print("開始監控 pending_books...")
    iteration = 0
    
    while iteration < max_iterations:
        try:
            # 獲取待處理書籍
            new_books = supabase.table('pending_books')\
                .select('isbn, title')\
                .eq('status', '待審核')\
                .is_('title', 'null')\
                .execute()

            if new_books.data:
                for book in new_books.data:
                    print(f"處理 ISBN: {book['isbn']}")
                    if update_pending_book(book['isbn']):
                        print(f"成功更新 ISBN {book['isbn']}")
                    else:
                        print(f"無法更新 ISBN {book['isbn']}")
            else:
                print("目前沒有待處理的書籍")
            
            iteration += 1
            print(f"第 {iteration}/{max_iterations} 次檢查完成")
            time.sleep(5)  # 每 5 秒檢查一次
            clear_output(wait=True)  # 清空輸出，保持乾淨
            
        except Exception as e:
            print(f"監控過程發生錯誤: {str(e)}")
            time.sleep(10)
            clear_output(wait=True)
    
    print("監控結束")

# 執行監控
monitor_pending_books(max_iterations=5)
```

```python
import requests
from bs4 import BeautifulSoup
import time
import os
import re

# 設定儲存書封的資料夾
STATIC_DIR = "static/covers"
os.makedirs(STATIC_DIR, exist_ok=True)

def crawl_book_info(isbn):
    search_url = f"https://search.books.com.tw/search/query/key/{isbn}/cat/all"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }
    try:
        # 搜尋頁面
        search_response = requests.get(search_url, headers=headers, timeout=5)
        if search_response.status_code != 200:
            print(f"搜尋 ISBN {isbn} 失敗，狀態碼: {search_response.status_code}")
            return None
        
        search_soup = BeautifulSoup(search_response.text, 'html.parser')
        product_link_elem = search_soup.select_one('a[href*="/redirect/move/"]')
        if not product_link_elem:
            print(f"ISBN {isbn} 未找到商品頁面")
            return None
        
        redirect_url = product_link_elem['href']
        if "item/" in redirect_url:
            item_id = redirect_url.split('item/')[1].split('/')[0]
            product_url = f"https://www.books.com.tw/products/{item_id}"
        else:
            print(f"無法從 redirect URL 解析商品 ID: {redirect_url}")
            return None

        # 商品頁面
        response = requests.get(product_url, headers=headers, timeout=5)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            
            # 提取書名
            title = soup.select_one('h1').text.strip() if soup.select_one('h1') else '未知書名'

            # 提取作者（先清理再提取）
            author_elem = soup.select_one('.type02_p003 li')
            author = '未知作者'
            if author_elem:
                author_text = author_elem.text.strip()
                print(f"DEBUG: 作者原始文字: {author_text}")
                # 移除「已追蹤作者：」到下一個「作者：」或結尾的所有內容
                author_text = re.sub(r'已追蹤作者：.*?作者：', '作者：', author_text, flags=re.DOTALL)
                # 如果沒有後續的「作者：」，移除「已追蹤作者：」到結尾
                author_text = re.sub(r'已追蹤作者：.*$', '', author_text, flags=re.DOTALL)
                print(f"DEBUG: 移除已追蹤後文字: {author_text}")
                # 提取「作者：」後的第一個名字
                match = re.search(r'作者：\s*([^\s]+)', author_text)
                if match:
                    author = match.group(1)
                print(f"DEBUG: 解析後作者: {author}")

            # 提取描述
            desc_elem = soup.select_one('.content')
            description = desc_elem.text.strip() if desc_elem else '無描述'

            # 提取書封圖片
            cover_elem = soup.select_one('img.cover')
            cover_url = cover_elem['src'] if cover_elem else None
            print(f"DEBUG: 原始書封 URL: {cover_url}")
            cover_data = None
            if cover_url:
                if not cover_url.startswith('http'):
                    cover_url = "https:" + cover_url
                print(f"DEBUG: 下載書封 URL: {cover_url}")
                cover_response = requests.get(cover_url, headers=headers, timeout=10)
                if cover_response.status_code == 200:
                    cover_data = cover_response.content
                else:
                    print(f"DEBUG: 書封下載失敗，狀態碼: {cover_response.status_code}")

            return {
                'title': title,
                'author': author,
                'description': description,
                'product_link': product_url,
                'cover_data': cover_data
            }
        else:
            print(f"訪問 ISBN {isbn} 商品頁面失敗，狀態碼: {response.status_code}")
            return None
    except requests.exceptions.RequestException as e:
        print(f"網路錯誤爬取 ISBN {isbn}: {str(e)}")
        return None
    except Exception as e:
        print(f"其他錯誤爬取 ISBN {isbn}: {str(e)}")
        return None

def save_book_info(isbn):
    info = crawl_book_info(isbn)
    if info:
        # 儲存書封
        cover_path = f"/static/covers/{isbn}.jpg"
        local_cover_path = os.path.join(STATIC_DIR, f"{isbn}.jpg")
        if info['cover_data']:
            with open(local_cover_path, 'wb') as f:
                f.write(info['cover_data'])
            print(f"書封已存到 {local_cover_path}")
        else:
            print(f"ISBN {isbn} 無書封")
        
        # 印出資訊
        print(f"書名: {info['title']}")
        print(f"作者: {info['author']}")
        print(f"描述: {info['description']}")
        print(f"商品連結: {info['product_link']}")
        return True
    return False

def test_scraper():
    test_isbn = "9786267555835"
    print(f"測試爬取 ISBN: {test_isbn}")
    if save_book_info(test_isbn):
        print(f"成功處理 ISBN {test_isbn}")
    else:
        print(f"無法處理 ISBN {test_isbn}")

if __name__ == "__main__":
    test_scraper()
```
